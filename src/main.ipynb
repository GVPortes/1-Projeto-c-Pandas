```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('../dados7.csv')
```

```python
#exibindo a tabela
display(df)
```

```python
#Aqui substitui os valores nulos que foram chamados "MD" pelo valor NaN
df.replace('MD', pd.NA)
```

```python
#Removendo os valores duplicados e a coluna Performance
df = df.drop_duplicates()
df = df.drop(['Performance'], axis=1)
```

```python
#usei isso para verificar quantas linhas e colunas ficaram após a  remoção
df.shape
```

```python
#aqui fiz um resumo analítico da coluna Domain Skills 1. Analisei a média, desvio padrão, valor mínimo e valor máximo.

df['Domain Skills 1'].describe()
```

```python
#Aqui eu verifico quantos valores iguais tem nessa coluna, assim descubro a Moda(valor que se repete por mais vezes nessa coluna).
display(df['Domain Skills 1'].value_counts())
```

```python
#Verificando o tipo de variavel das colunas
print(df['Year of Birth'].dtype)
print(df['Domain Skills 1'].dtype)
```

```python
#criando um gráfico analisando a Domain Skills 1 relacionada a coluna Gender:
#já verifiquei e a coluna domain skills 1 é toda float entao nao preciso converter ela
#abaixo estamos associando o Gender com a Domain Skills 1
GenderXSkill = df.groupby('Year of Birth')['Domain Skills 1'].mean()
#criando o gráfico:
plt.figure(figsize=(10, 5))
GenderXSkill.plot(kind='bar', color=['blue', 'red'])
plt.title('Média de Domain Skills 1 por Year of Birth')
plt.show()
```

```python
#Verificando o tipo de variavel das colunas
print(df['Domain Skills 2'].dtype)
print(df['Specialization in study'].dtype)
```

```python

#Criando um gráfico que analisa Specialization in study relacionado a Domain Skill 2
DomainSkills2_SpecializationInStudy = df.groupby('Specialization in study')['Domain Skills 2'].mean()
#criando o gráfico:
plt.figure(figsize=(10, 5))
DomainSkills2_SpecializationInStudy.plot(kind='bar', color=['blue', 'red'])
plt.title('Média de Domain Skills 2 por Specialization in study')
plt.show()
```

```python
out_of_bounds_10th = df[(df[' 10th percentage'] < 0) | (df[' 10th percentage'] > 100)]
out_of_bounds_12th = df[(df[' 12th percentage'] < 0) | (df[' 12th percentage'] > 100)]
```

```python

print("Valores fora da escala - 10th percentage:\n", out_of_bounds_10th)
print("Valores fora da escala - 12th percentage:\n", out_of_bounds_12th)
```

```python
# Verificar outliers em Domain Skills 1 usando o desvio padrão
mean = df['Domain Skills 1'].mean()
std_dev = df['Domain Skills 1'].std()

# Definir um limite de 3 desvios padrão
outliers = df[(df['Domain Skills 1'] < mean - 3 * std_dev) | (df['Domain Skills 1'] > mean + 3 * std_dev)]


# Você pode remover ou tratar esses outliers
df = df[~df.isin(outliers)]
```

```python
print("Outliers em Domain Skills 1:\n", outliers)
```

```python
#removendo dados que são irrelevantes para a análise, coluna "Number of characters in Original Name" e a coluna Name pois nao faz sentido analisar esses dados.
df = df.drop(['Number of characters in Original Name'], axis=1)
df = df.drop(['Name'], axis=1)
```

```python
#verificando os nomes das colunas existentes para ver se removi as colunas corretas

df.columns
```

```python
# Salvar o DataFrame com as alterações em um novo arquivo CSV, o False é para ele nao criar um index numerando linha a linha. Por acaso, aprendi isso enquanto escrevia esse trecho de código.
#ao fazer vi que ele tinha gerado com o indice, ai pesquisei como faria para ele nao gerar e achei o index=False.
df.to_csv('dados7_alterado.csv', index=False)
```
